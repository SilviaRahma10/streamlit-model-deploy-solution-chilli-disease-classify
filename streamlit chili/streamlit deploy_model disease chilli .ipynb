```python

%%writefile app.py
import streamlit as st
import pandas as pd
from PIL import Image
import streamlit as st


import streamlit as st
import pandas as pd
from PIL import Image, ImageOps
import matplotlib.pyplot as plt
import tensorflow as tf
import tensorflow_hub as hub
import numpy as np
from tensorflow.keras import preprocessing
from tensorflow.keras.models import load_model
from tensorflow.keras.activations import softmax
import os
import h5py


def main_page():
    st.sidebar.markdown("###### CHILLI DISEASE PREDICTION")
    with st.container():
        st.markdown('---')
        st.markdown\
             ("<h1 style='text-align: center; color: #728FCE;'>CHILLI DISEASE PREDICTION</h1>", unsafe_allow_html=True)
        st.markdown("<h3 style='text-align: center; color: #728FCE;'>(by Silvia Rahma)</h1>", unsafe_allow_html=True)
        st.markdown("<h1 style='text-align: center; color: #728FCE;'>🦠⚠️🖥️</h1>", unsafe_allow_html=True)
        st.markdown('---')
        st.markdown("<h3 style='text-align: center; color: #728FCE;'>Features</h3>", unsafe_allow_html=True)
        st.markdown('- Disease Prediction'
            '\n- Disease Definition'
            '\n- Disease Occurence'
            '\n- Disease Importance'
            '\n- Disease Treatment')


# Cabai
def chilli():
    st.sidebar.markdown('- Deteksi penyakit Cabai'
                    '\n- Penjelasan Penyakit Cabai')
    st.markdown('---')
    st.markdown("<h1 style='text-align: center; color: #728FCE;'>Tanaman Cabai</h1>", unsafe_allow_html=True)
    st.markdown("<h1 style='text-align: center; color: #728FCE;'>🦠</h1>", unsafe_allow_html=True)
    st.markdown('---')

    # Deteksi penyakit
    st.markdown("<h2 style='color: #728FCE;'>Upload Gambar Daun Cabai</h1>", unsafe_allow_html=True)

        #fungsi untuk prediksi kelas
    def predict_class_cabai(image):
        classifier_model = tf.keras.models.load_model(r'/content/drive/MyDrive/portofolio/streamlit chili/Resnet50Model_chili_v1.hdf5')
        shape = ((224, 224, 3))
        model = tf.keras.Sequential([hub.KerasLayer(classifier_model, input_shape=shape)])

        test_image = image.resize((224, 224))
        test_image = preprocessing.image.img_to_array(test_image)
        #test_image = test_image/225.0
        test_image = np.expand_dims(test_image, axis = 0)
        class_names = ['chili  leaf spot', 'chili  whitefly', 'chili leaf curl', 'chili yellowish']
        prediction = model.predict(test_image)
        score = tf.nn.softmax(prediction[0])
        score = score.numpy()
        image_class =  class_names[np.argmax(score)]
        return image_class

 # fungsi untuk import atau upload gambar
    file_uploaded = st.file_uploader("Pilih File", type = ['jpg', 'png', 'jpeg'])
    if file_uploaded is not None:
        image = Image.open(file_uploaded)
        figure = plt.figure(figsize=(4, 3))
        plt.imshow(image)
        plt.axis('off')
        predict_class = predict_class_cabai(image)
        st.markdown("<h2 style='color: #728FCE;'>Prediksi penyakit Cabai : {}</h1>".format(predict_class), unsafe_allow_html=True)
        st.pyplot(figure)

        # Penjelasan Penyakit
        st.markdown("<h2 style='color: #728FCE;'>Penjelasan Penyakit : {}</h1>".format(predict_class), unsafe_allow_html=True)
        if(predict_class=='chili  leaf spot'):
            st.write('#### Definisi :')
            st.write(" Leaf Spot atau bercak daun pada cabai merupakan penyakit yang disebabkan oleh jamur patogen cercospora capsici. ")
            st.write('#### Kemunculan :')
            st.write(" Penyakit leaf spot pada cabai sering terjadi pada daerah dengan suhu rata-rata 22,5-23,5 celcius dan dengan kelembaban  relatif 77- 85 %. Jamur dapat menyebar melalui air hujan, angin, peralatan pertanian, atau bahkan serangga.")
            st.write('#### Penting:')
            st.write(" Penyakit bercak daun cabe dengan tingkat keparahan yang tinggi dapat menyebabkan daun menjadi rontok dan hingga pembusukan pada ujung batang serta membuat cabai terkena sunscald. ")
            st.write('#### Penyembuhan:')
            st.write(" - Rendam benih cabe pada air panas dengan suhu 52 derajat celcius selama 30 menit sebelum disemai atau ditanaman. Semprotkan agen biokontrol Trichoderma 5-10 gm/liter air atau trichoderma @5-10 ml/liter air ")
            st.write(" - Lakukan sanitasi pada area tanaman dan hindari penumpukan sisa tanaman yang terinfeksi di sekitar tanaman cabai. Lakukan rotasi tanaman dengan menghindari penanaman cabai atau anggota keluarga tanaman cabai di lahan yang sama setiap tahun untuk mengurangi penumpukan patogen di tanah. Hindari irigasi berlebihan dan buat drainase yang baik. ")
            st.write(" - Lakukan penanaman bergiliran,yaitu setelah ditanami kentang maka untuk berikutnya ditanami palawija seperti jagung, atau kedela ")
            st.write(" - Lakukan Solarisasi tanah untuk mematikan bibit penyakit")
            st.write('#### Fungisida :')
            st.write(" - WUZ 200/125")

        elif(predict_class=='chili  whitefly'):
            st.write('#### Definisi :')
            st.write(" Whitefly merupakan penyakit pada tanaman cabai yang disebabkan oleh hama serangga kecil (Bemisia tabaci). Selain itu serangga Bemisia tabaci ini juga dapat menularkan virus lain pada tanaman yang dihinggapinya.")
            st.write('#### Kemunculan :')
            st.write(" Whitefly cenderung berkembang dengan baik pada suhu yang hangat (sekitar 20-30 derajat Celcius) dan kelembaban yang tinggi")
            st.write('#### Penting:')
            st.write(" Penyakit whitefly berdampak pada melambatkan pertumbuhan tanaman, kerusakan pada daun dan penurunan hasil panen")
            st.write('#### Insektisida :')
            st.write(" - Alika 247 ZC ")

        elif(predict_class=='chili leaf curl'):
            st.write('#### Definisi :')
            st.write(" Leaf curl atau daun keriting merupakan penyakit pada tanaman cabai yang disebabkan oleh adanya virus chili leaf curl virus atau ChiLCV yang ditandai dengan dengan daun yang melipat, mengerut ataupun menggulung.")
            st.write('#### Kemunculan :')
            st.write(" Penyakit leaf curl mengalami penyebaran yang cepat pada daerah dengan kelembaban yang tinggi dan suhu yang rendah,serta perubahan cuaca yang ekstrem. Penyakit ini dapat menyebar melalui serangga, peralatan pertanian yang terkontaminasi atau melalui kontak langsung dengan tanaman yang terinfeksi virus ChiLCV.  ")
            st.write('#### Penting:')
            st.write(" Penyakit leaf curl pada akan berdampak pada pertumbuhan cabai, seperti ukuran buah mengecil, cacat, penurunan jumlah produksi bahkan tidak menghasilkan buah sama sekali.")
            st.write('#### Penyembuhan:')
            st.write(" - Pilih varietas yang tidak terlalu rentan terhadap penyakit leaf curl untuk mengurangi resiko infeksi seperti Catas, Bara, Genie, Raya, Sonar. Kultur Teknis ")
            st.write(" - Lakukan pembasmian pada tanaman cabai yang terinfeksi penyakit leaf curl untuk mencegah penyebaran ke tanaman cabai lainnya. Jaga kebersihan peralatan pertanian dengan membersihkannya secara teratur. Pastikan pisau pemangkas, gunting, atau alat lainnya yang digunakan pada tanaman yang terinfeksi penyakit leaf curl tidak digunakan kembali pada tanaman sehat tanpa dicuci dan disterilkan terlebih dahulu.")
            st.write(" - Lakukan pemangkasan pada daun yang terinfeksi leaf curl, dan pada tanaman cabai yang terlalu padat agar sirkulasi udara disekitar tanaman tetap stabil.")
            st.write(" - Jaga kebersihan peralatan pertanian dengan membersihkannya secara teratur. Pastikan pisau pemangkas, gunting, atau alat lainnya yang digunakan pada tanaman yang terinfeksi penyakit leaf curl tidak digunakan kembali pada tanaman sehat tanpa dicuci dan disterilkan terlebih dahulu.")
            st.write('#### Insektisida :')
            st.write(" - insektisida thioxam ")

        elif(predict_class=='chili yellowish'):
            st.write('#### Definisi :')
            st.write(" Yellowish merupakan penyakit pada tanaman cabai yang yang disebabkan oleh adanya Virus Yellow Leaf Curl (TYLCV) yang ditularkan oleh serangga sehingga dapat menginfeksi daun tanaman. ")
            st.write('#### Kemunculan :')
            st.write(" Penyakit yellowish pada cabai ditularkan melalui serangga vektor kutu kebul dan bukan termasuk seed born disease atau tidak ditularkan melalui benih. Importance : Penyakit yellowish dapat menurunkan hasil panen hingga 75% terutama saat musim kemarau. ")
            st.write('#### Penting:')
            st.write(" Penyakit yellowish dapat menurunkan hasil panen hingga 75% terutama saat musim kemarau.")
            st.write('#### Penyembuhan:')
            st.write(" - Menanam jenis verietas cabai yang tahar terhadap virus kuning yaitu varietas Cabe Kopay, dan lakukan perendaman benih dengan menggunakan Plant Growth Promoting Rhizobacteria")
            st.write('#### Produk:')
            st.write(" - Plant Growth Promoting Rhizobacteria")



page_names_to_funcs = {
    "Headline": main_page,
    "Deteksi Penyakit Tanaman Cabai": chilli,
}

selected_page = st.sidebar.selectbox("Select a page", page_names_to_funcs.keys())
page_names_to_funcs[selected_page]()
```
